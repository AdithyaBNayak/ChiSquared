```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix1.csv')
df
```

```python
df.shape
```

```python
# Remove the Duplicates
df.drop_duplicates()
```

```python
# Adding a new column 
df["no_of_seasons"] = 0
df
```

```python
for row in df.index:
    if df.loc[row,'type'] == 'TV Show':
        df.loc[row, 'no_of_seasons'] = df.loc[row, 'duration']
        df.loc[row, 'duration'] = np.nan
    else:
        df.loc[row, 'no_of_seasons'] = np.nan
        

df
```

```python
#Formatting Dates in the same structure
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].dt.strftime('%Y-%m-%d')
df
```

```python
# Directors should be NaN if not given
df['director'] = df['director'].replace("Not Given", "NaN")
df
```

```python
# Remove "Season"/"Seasons" from the column
df['no_of_seasons'] = df['no_of_seasons'].str.replace(' Seasons', '')
df['no_of_seasons'] = df['no_of_seasons'].str.replace(' Season', '')
df
```

```python
# Remove "min"" from the duration column
df.rename(columns={'duration': 'duration_in_mins'}, inplace=True)
df['duration_in_mins'] = df['duration_in_mins'].str.replace(' min', '')
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv("generated_nextflix.csv", index=False)
```
