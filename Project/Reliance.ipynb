```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# Extract year from the 'Date' column and group by year
df['Year'] = df['Date'].dt.year
df
```

```python
yearly_df= df.groupby('Year').agg({'Open': 'mean', 'Close': 'mean'}).reset_index()
yearly_df
```

```python
plt.figure(figsize=(12,6))
plt.plot(yearly_df['Year'], yearly_df['Open'])
plt.plot(yearly_df['Year'], yearly_df['Close'])
plt.xlabel('Year')
plt.ylabel('Open/Closing Price Avg')
```
